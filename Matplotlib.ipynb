# Matplotlib

Matplotlib is the most popular plotting library for Python

```python
import numpy as np

#Library used to plot graphs,bars etc for Statical calculations and much more
import matplotlib.pyplot as plt

#this allows to show graphd in Jupyter Notebook without using show() function
%matplotlib inline
```

# Function Approach

```python
x = np.linspace(0,5,11)  #defining X value  for future graph building
y = x ** 2  # defining Y value for future graph building
```

```python
print(x)
len(x)
```

```python
print(y)
len(x)
```

```python
#plotting a simple graph
plt.plot(x,y,'y')
plt.xlabel('X Axis')    #Label for X axis
plt.ylabel('Y Axis')   #Label for Y axis
plt.title("Graph for X = Y .....")   #Title for the given graph

```

# Object Oriented 

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel("X Values")
axes.set_ylabel("Y Values")
axes.set_title("Graph designed by OOPS methos......")
```

```python

fig1 = plt.figure()   #creates the layout for the given graph
axes = fig1.add_axes([0.1,0.1,0.8,0.8])    #set the axis for the given graph....(left,bottom,width,height)
axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')

axes.set_title('Graph Plotting ......')
```

```python
#saving the figure as PNG formate by the name OOPS Graph with pixel per inch
fig1.savefig('OOPS Graph.png',dpi=100)
```

```python
fig2 = plt.figure()
axes1 = fig2.add_axes([0.1,0.4,0.6,0.8])

axes2 = fig2.add_axes([0.2,0.9,0.3,0.2])

axes1.plot(x,y)
#axes1.show()

axes2.plot(y,x)
#axes2.show()
fig2.savefig('graph with in graph.jpg',dpi=100)
```

```python
fig3 = plt.figure(figsize=(6,3), dpi = 100)

ax = fig3.add_axes([0,0,1,1])

ax = axes.plot(x,y,'g.')
```

# Subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,3)
plt.plot(y,x,'r')

plt.tight_layout()
#fig.savefig('my_graph.jpg', dpi=100)
```

```python

plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Subplot')

print('\t')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
plt.xlabel('X Label')
#plt.ylabel('Y Label')
plt.title('Subplot')
```

```python
#fig4 = plt.figure()
fig4 , axes3 = plt.subplots(1,3)

color=['r','g','b']
for i, ax in enumerate(axes3):
    ax.plot(x,y, color[i])
    
fig4.tight_layout()
```

```python

fig4.savefig('Subplots.png',dpi=100)
```

```python
fig5,ax = plt.subplots(1,3 ,figsize = (3,2), dpi = 100)

#ax = fig5.add_axes([0,0,1,1])
ax[0].plot(x,y)
ax[1].plot(y,x)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, label ='x = y''ro--')
plt.legend()
plt.subplot(1,2,2)
plt.plot(x, x**4, 'b.--', label = 'x = y**4')
plt.legend()
```
