# Seaborn
Seaborn is a Python data visualization library based on matplotlib. 
It provides a high-level interface for drawing attractive and informative statistical graphics
It is built on top of matplotlib and closely integrated with pandas data structures.

Here is some of the functionality that seaborn offers:

A dataset-oriented API for examining relationships between multiple variables

Specialized support for using categorical variables to show observations or aggregate statistics

Options for visualizing univariate or bivariate distributions and for comparing them between subsets of data

Automatic estimation and plotting of linear regression models for different kinds dependent variables

Convenient views onto the overall structure of complex datasets

High-level abstractions for structuring multi-plot grids that let you easily build complex visualizations

Concise control over matplotlib figure styling with several built-in themes

Tools for choosing color palettes that faithfully reveal patterns in your data

Seaborn aims to make visualization a central part of exploring and understanding data.
Its dataset-oriented plotting functions operate on dataframes and arrays containing whole datasets and internally
perform the necessary semantic mapping and statistical aggregation to produce informative plots

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
sns.set(style = 'darkgrid')

#.... The Seaborn has many DataSets inbuilt, which we can use for visualization and prctice
#.... tips is one of the inbuilt dataset of the seaborn 

tips = sns.load_dataset('tips') #... loads the dataset for data visualization
```

```python
tips.head() #....prints the head elements of the dataset i-e first five rows of dataset
```

```python
sns.relplot(x='total_bill', y='tip', data=tips) 
#...it draws the plot between total_bill and tip values from the tips dataset present in seaborn
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'smoker')
```

```python
tips['smoker'].value_counts()  #... prints the no of smokers and non smokers in a dataset
```

```python
sns.relplot(x='total_bill', y='tip',hue = 'size', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip',hue = 'size', data=tips, palette = 'ch:r= -0.5, l=0.75')

#...palette provised the colour for the datd point in the plot printed
```

```python
sns.relplot(x='total_bill', y='tip', size= 'size', data=tips)

#... the size='size' will increase the size of the ploted data points with the increase of date in the dataset
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, style = 'size')

#.... style='size' will plot the points as some symbols like printed in below plot 
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, size='size', sizes = (50,500))
#...Sizes = (_ , _) will normalize each of the parameter
```

# Line plottng 

In the above portion there were by defaut the scatter type plots generated,
but in this portion we use the line 
to plot the plots by using the " kind = 'line' " to represent plots 

```python
from numpy.random import randn
```

```python
 #... create the dataframe df 
    
df = pd.DataFrame(dict(time = np.arange(500),value = randn(500).cumsum()))
df.head()
```

```python
#....seaborn plot for the above created dataframe(dataset) with kind = line

#....it is by default sorted .....

sns.relplot(x = 'time', y = 'value', kind ='line', data = df)
```

```python
#..... here we have created unsorted dataframe 


df = pd.DataFrame(randn(500,2).cumsum(axis =0), columns = ['time','value'])
df.head()
```

```python
#.......we sort it mannualy by using "sort=True"


sns.relplot(x = 'time', y = 'value',kind = 'line', data = df, sort = True)
```

```python
#.... here we declear "sort=False" 

sns.relplot(x = 'time', y = 'value',kind = 'line', data = df, sort = False)
```

```python
#....Load another Dataset "fmri"
#.... fmri is also one of the inbuilt dataset of Seaborn 

fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line' ,data = fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line' ,data = fmri, ci = False)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line' ,data = fmri, ci = 'sd')

#.... ci stands for confedence interval ci =False will remove the shaded area fom the plot like from the above plot.....
#.... ci = sd will plot standard devation along with line lin in this plot
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line' ,data = fmri, estimator = None)

#.... estimator will plot the real view of the plot
```

```python
sns.relplot(x = 'timepoint', y = 'signal',hue = 'region',style='event', kind = 'line' ,data = fmri, marker = True)
```

```python
sns.relplot(x = 'timepoint', y = 'signal',hue = 'region', style = 'event', kind = 'line' ,data = fmri,markers = True)
```

```python
#.... Load another dataset namelt dots...

dots = sns.load_dataset('dots')
dots.head()
```

```python
#..... we can also select dataset and make query on it for selected enteties

dots = sns.load_dataset('dots').query("align == 'dots' ")

dots.head()
```

```python
sns.relplot(x = 'time', y = 'firing_rate',data = dots, kind = 'line', hue = 'coherence', style = 'choice')
```

```python
#.... Create manually colors for the lines in the plot
#.... cubehelix_palette is a predefined function for this color formation
#.... n_colors is the number of colors to specify for the pllot

palette = sns.cubehelix_palette(light = 0.8,n_colors = 6)
```

```python
sns.relplot(x = 'time', y = 'firing_rate',data = dots, kind = 'line',hue = 'coherence', style = 'choice', palette = palette)
```
